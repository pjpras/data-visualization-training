```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('../house_prices.csv')

# Step 2: Handle missing data (if any)
df = df.dropna(subset=['price'])

# Step 3: Create a figure and subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Step 4: Plot a histogram of house prices
axs[0].hist(df['price'], bins=10, color='blue', edgecolor='black')
axs[0].set_title('Histogram of House Prices')
axs[0].set_xlabel('Price')
axs[0].set_ylabel('Frequency')

# Step 5: Plot a pie chart of house prices
price_ranges = pd.cut(df['price'], bins=10)
price_range_counts = price_ranges.value_counts().sort_index()
axs[1].pie(price_range_counts, labels=[f'{interval.left:.0f}-{interval.right:.0f}' for interval in price_range_counts.index], autopct='%1.1f%%', startangle=140)
axs[1].set_title('Pie Chart of House Prices')

# Step 6: Plot a line plot of house prices
axs[2].plot(df['price'].reset_index(drop=True), marker='o', linestyle='-', color='green')
axs[2].set_title('Line Plot of House Prices')
axs[2].set_xlabel('Index')
axs[2].set_ylabel('Price')

# Adjust layout to add space between plots
plt.tight_layout()
plt.show()
```
